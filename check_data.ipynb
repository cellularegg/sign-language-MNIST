```python
# import general utilities
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import cv2
import imutils
```

```python
# read the data
train_df = pd.read_csv("data/sign_mnist_train.csv")
test_df = pd.read_csv("data/sign_mnist_test.csv")

display(train_df)
```

```python
# extract labels and prepare data
train_label = train_df['label'].values
train_data = []
for i in train_df.drop('label', axis=1).to_numpy():
    train_data.append( i.reshape(28,28) )
train_data = np.array(train_data)

test_label = test_df['label'].values
test_data = []
for i in test_df.drop('label', axis=1).to_numpy():
    test_data.append( i.reshape(28,28) )
test_data = np.array(test_data)
```
